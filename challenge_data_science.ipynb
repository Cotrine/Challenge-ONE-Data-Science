```python
import pandas as pd
```

```python
url1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv'
url2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv'
url3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv'
url4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv'
```

```python
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
df4 = pd.read_csv(url4)
```

```python
tiendas = [df1,df2,df3,df4]
```

##**ingreso total por cada tienda**

```python
for i, df in enumerate(tiendas, 1):
  total=df['Precio'].sum()
  print(total)
```

##**valoracion media por tienda**

```python
for i, df in enumerate(tiendas, 1):
    promedio = df['Precio'].mean()
    print(f"Tienda {i} - Promedio de precios: ${promedio:,.2f}")
```

```python
for i, df in enumerate(tiendas, 1):
    promedio = df['Costo de envío'].mean()
    print(f"Tienda {i} - valor de envio promedio: ${promedio:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [df1['Precio'].sum(), df2['Precio'].sum(), df3['Precio'].sum(), df4['Precio'].sum()]
plt.bar(nombres, ingresos, color=['#4e79a7', '#f28e2b', '#e15759', '#76b7b2'])
plt.title('Ingreso Total por Tienda')
plt.ylabel('Suma de Precios ($)')
plt.tight_layout()
plt.show()
```

```python
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [df1['Precio'].sum(), df2['Precio'].sum(), df3['Precio'].sum(), df4['Precio'].sum()]

plt.subplot(1, 3, 2)
plt.pie(ingresos, labels=nombres, autopct='%1.1f%%', startangle=140)
plt.title('Proporción de Ventas')

plt.tight_layout()
plt.show()
```

```python
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [df1['Precio'].sum(), df2['Precio'].sum(), df3['Precio'].sum(), df4['Precio'].sum()]

plt.boxplot([df1['Precio'], df2['Precio'], df3['Precio'], df4['Precio']], labels=nombres)
plt.title('Distribución de Precios')
plt.ylabel('Precio ($)')

plt.tight_layout()
plt.show()
```
